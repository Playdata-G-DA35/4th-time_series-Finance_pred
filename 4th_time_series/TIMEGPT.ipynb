```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
data = {
    'USD/KRW예측값': [1374.764526, 1377.674072, 1380.771484, 1381.440674, 1379.498291],
    'USD/KRW실제값': [1368.73999, 1365.25, 1379.050049, 1373.959961, 1377.359985],
}

df = pd.DataFrame(data, index=['6/6', '6/7', '6/8', '6/9', '6/10'])

# '공휴일' 부분은 실제 값이 없어 비교 불가능하므로 제거
df.dropna(inplace=True)

# 실제값과 예측값만 추출하여 numpy array로 변환
y_true_usd_krw = df['USD/KRW실제값'].values.astype(float)
y_pred_usd_krw = df['USD/KRW예측값'].values.astype(float)

# MSE와 MAE 계산
mse_usd_krw = mean_squared_error(y_true_usd_krw, y_pred_usd_krw)
mae_usd_krw = mean_absolute_error(y_true_usd_krw, y_pred_usd_krw)

# 결과 출력
print("USD/KRW 예측값과 실제값:")
print("Mean Squared Error (MSE):", mse_usd_krw)
print("Mean Absolute Error (MAE):", mae_usd_krw)
```
